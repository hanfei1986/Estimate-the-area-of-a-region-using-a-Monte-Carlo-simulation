## Calculate the area of the grey region
### Method 1

```python
import numpy as np
```

```python
throws = 100000
x_lower = 0
x_upper = 2
y_lower = 0
y_upper = 2
hits = 0
```

```python
for i in range(throws):
    # Choose random X and Y
    x = np.random.uniform(x_lower, x_upper)
    y = np.random.uniform(y_lower, y_upper)
    # If the point is inside the grey region, increase variable
    if x**2+y**2 >= 2**2 and (x-1)**2+(y-1)**2 <= 1**2:
        hits += 1
```

```python
# Calculate area
area = (x_upper-x_lower)*(y_upper-y_lower)*hits/throws
area
```

### Method 2

```python
X = np.random.uniform(x_lower, x_upper, throws)
Y = np.random.uniform(y_lower, y_upper, throws)
```

```python
res1 = X**2+Y**2
res2 = (X-1)**2+(Y-1)**2
```

```python
hits = np.logical_and((res1 >= 2**2), (res2 <= 1**2)).sum()
```

```python
# Calculate area
area = (x_upper-x_lower)*(y_upper-y_lower)*hits/throws
area
```
